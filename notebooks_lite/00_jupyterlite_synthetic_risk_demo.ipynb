# JupyterLite synthetic risk demo

This notebook is designed for browser-only execution. It does not require geopandas, rasterio, GDAL, or local GIS data.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

rng = np.random.default_rng(42)
n = 30
facilities = pd.DataFrame({
    "facility_id": [f"HF{i:03d}" for i in range(n)],
    "spillover_suitability": rng.beta(2, 3, n),
    "human_exposure": rng.beta(2.5, 2.5, n),
    "importation_pressure": rng.beta(1.5, 4, n),
    "facility_vulnerability": rng.beta(2, 2, n),
})
weights = {"spillover_suitability": 0.25, "human_exposure": 0.25, "importation_pressure": 0.30, "facility_vulnerability": 0.20}
facilities["risk_score"] = sum(facilities[k] * w for k, w in weights.items())
facilities["risk_tier"] = pd.qcut(facilities["risk_score"].rank(method="first"), 4, labels=["low", "moderate", "high", "very_high"])
facilities.sort_values("risk_score", ascending=False).head(10)
```

```python
top = facilities.sort_values("risk_score", ascending=False).head(15).iloc[::-1]
plt.figure(figsize=(8, 5))
plt.barh(top["facility_id"], top["risk_score"])
plt.xlabel("Risk score")
plt.title("Synthetic facility risk ranking")
plt.tight_layout()
plt.show()
```
